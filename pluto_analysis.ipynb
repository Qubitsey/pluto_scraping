```python
import pandas as pd
import matplotlib.pyplot as plt
# Cargar el archivo CSV
df = pd.read_csv('pluto_227s.csv')

# Mostrar las primeras filas de forma transpuesta para verificar la carga
print(df.head().T)
```

```python
# Función para identificar el tipo de contenido basado en la URL
def identify_content_type(url):
    if 'series' in url:
        return 'Serie'
    elif 'live-tv' in url:
        return 'Canal de TV en Vivo'
    else:
        return 'Película'

df['Content_Type'] = df['Movies, Series and Channels urls'].apply(identify_content_type)

content_type_counts = df['Content_Type'].value_counts()
print("Conteo de tipos de contenido:\n", content_type_counts)

# Gráfico de barras para los tipos de contenido
content_type_counts.plot(kind='bar', color='orange')
plt.title('Distribución de Tipos de Contenido')
plt.xlabel('Tipo de Contenido')
plt.ylabel('Frecuencia')
plt.show()
```
